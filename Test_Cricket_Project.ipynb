```python
#Q4 : Max highest inns score by country 
```

```python
df.groupby('Country')['Highest_Score'].max()
```

```python
max_scores = df.groupby('Country')['Highest_Score'].max()

# Create a new DataFrame with the maximum scores for each country
df_max_scores = pd.DataFrame({'Country': max_scores.index, 'Max_Highest_Score': max_scores.values})

# Sort the new DataFrame by 'Max_Highest_Score' in descending order
df_max_scores_sorted = df_max_scores.sort_values(by='Max_Highest_Score', ascending=False)

# Display the sorted DataFrame
print(df_max_scores_sorted)
```

```python
#Question 5 : hundreds, fifities, ducks(0) avg by country
```

```python
# Group by 'Country' and calculate the mean of '100' for each group
result = df.groupby('Country')['100'].mean()

# Sort the result in descending order
result_sorted = result.sort_values(ascending=False)

print(result_sorted)
```

```python
#AVG 50 by country
result = df.groupby('Country')['50'].mean()

result_sorted = result.sort_values(ascending=False)

print(result_sorted)
```

```python
result = df.groupby('Country')['0'].mean()

result_sorted = result.sort_values(ascending=False)

print(result_sorted)
```
